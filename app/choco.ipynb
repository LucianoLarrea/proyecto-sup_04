```python
import pandas as pd

df = pd.read_csv('../data/choco.csv', sep=',')
df.head()
```

```python
encabezados = list(df.columns)
print(encabezados)
```

```python
df.rename(columns={'Company\xa0\n(Maker-if known)':'company'}, inplace=True)
df.rename(columns={'Specific Bean Origin\nor Bar Name':'name-origin'}, inplace=True)
df.rename(columns={'REF':'ref'}, inplace=True)
df.rename(columns={'Review\nDate':'review'}, inplace=True)
df.rename(columns={'Cocoa\nPercent':'cacao'}, inplace=True)
df.rename(columns={'Company\nLocation':'location'}, inplace=True)
df.rename(columns={'Rating':'rating'}, inplace=True)
df.rename(columns={'Bean\nType':'bean'}, inplace=True)
df.rename(columns={'Broad Bean\nOrigin':'bean-origin'}, inplace=True)
df.info()
```

```python
# Elimina el signo % y convertir la columna 'cacao' a formato numérico
df['porcentaje'] = df['cacao'].str.replace('%', '').astype(float)

# Define una función lambda para asignar valores
def assign_pur(cacao):
    if cacao < 60:
        return 'BAJO'
    elif cacao < 80:
        return 'INTERMEDIO'
    elif cacao < 100:
        return 'ALTO'
    else:
        return 'PURO'

# Aplica la función lambda a la columna 'cacao' y guarda el resultado en la nueva columna 'pur'
df['pur'] = df['porcentaje'].apply(lambda x: assign_pur(x))
df.head()
```

```python
# Eliminar la columna 'porcentaje'
# df.drop(columns=['porcentaje'], inplace=True)
# Definir función lambda para concatenar valores
df['res'] = df.apply(lambda x: f"{x['company']}&{x['pur']}&{x['rating']}&{x['location']}", axis=1)
df.nunique()
```

```python
# Guardar cambios en un archivo CSV
df.to_csv('chocobar.csv', index=False)
```
